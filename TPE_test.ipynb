```python
import numpy as np
import random
import cv2

#임의 난수 생성(Pseudo Random Function)
def PRF(b):
    random_PI = np.array([])
    for x in range(b*b):
        random_PI = np.append(random_PI, np.array([random.randrange(b*b - x)]))
    return random_PI

#순열 함수
def purmutation(image):
    i, j = image.shape[0], image.shape[1]
    i = int(i/b)
    j = int(j/b)
    random_PI = np.array([])
    # PI = PRF(b)
    for w in range(3):
        for z in range(j):
            for y in range(i):
                block = image[y*b:(y+1)*b,z*b:(z+1)*b,w]
                block_list = block.flatten()
                arr = np.array([])
                random_PI_block = PRF(b)
                for n in range(b*b):
                    pi = random_PI_block[n]
                    # pi = PI[n]
                    arr = np.append(arr, block_list[int(pi)])
                    m = 0
                    while(int(pi) + m + 1 != b*b - n):
                        block_list[int(pi) + m] = block_list[int(pi) + m + 1]
                        m += 1
                image[y*b:(y+1)*b,z*b:(z+1)*b,w] = arr.reshape(b,b)
                random_PI = np.append(random_PI, random_PI_block)
    return random_PI
#순열 역함수
def inverse_purmutation(image, random_PI):
    i, j = image.shape[0], image.shape[1]
    i = int(i/b)
    j = int(j/b)
    for w in range(3):
        for z in range(j):
            for y in range(i):
                block = image[y*b:(y+1)*b,z*b:(z+1)*b,w]
                block_list = block.flatten()
                arr = np.array([])
                random_PI_block = random_PI[(y + (i*z) + (i*j*w)) * b*b : (y + (i*z) + (i*j*w)) * b*b + b*b]
                for n in range(b*b):
                    pi = random_PI_block[b*b - n - 1]
                    # pi = PI[b*b - n - 1]
                    arr = np.append(arr, np.array([0]))
                    m = 0
                    while(n - m > int(pi)):
                        arr[n - m] = arr [n - m - 1]
                        m += 1
                    arr[int(pi)] = block_list[b*b - n - 1]
                image[y*b:(y+1)*b,z*b:(z+1)*b,w] = arr.reshape(b,b)

#대체 함수,,,sum이 d(255)보다 크면 다른 픽셀값으로 대체함
def substitution(image):
    i, j = image1.shape[0], image1.shape[1]
    for z in range(3):
        for y in range(j):
            for x in range(0,i-1,2):
                a = image[x,y,z]
                b = image[x+1,y,z]
                if (a + b) > 255:
                    image[x,y,z] = 255 - a
                    image[x+1,y,z] = 2*a + b - 255

#이미지 불러오기
image1 = cv2.imread('C:/Users/Park Jiwon/source2/test2.jpg')
cv2.imshow('image1',image1)

# 블럭 크기 b x b
b = 20

#round 만큼 순열, 대체 반복
for r in range(1):
    substitution(image1)
    random_PI = purmutation(image1)

#암호화 이미지 출력
cv2.imshow('image2',image1)

#암호화 이미지 저장
# save_file = "C:/Users/Park Jiwon/source2/test2_encyption.jpg"
# cv2.imwrite(save_file, image1)

#복원
for r in range(1):
    inverse_purmutation(image1, random_PI)
    substitution(image1)

#복원 이미지 출력
cv2.imshow('image3',image1)

# #썸네일
# for w in range(3):
#     for z in range(image1.shape[1]):
#         for y in range(image1.shape[0]):
#             image1[y*b:(y+1)*b,z*b:(z+1)*b,w] = image1[y*b:(y+1)*b,z*b:(z+1)*b,w].sum() / (b*b)
# image_resize = cv2.resize(image1,dsize=(0,0),fx=0.5,fy=0.5)

# cv2.imshow('image4',image_resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
